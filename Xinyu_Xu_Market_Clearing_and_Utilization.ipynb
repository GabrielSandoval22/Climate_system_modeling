```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
REFERENCE_BUSBAR_ELEC_PRICE = 24.4  # Reference busbar electricity price
PRICE_ADJUSTMENT_TIME = 0.5  # Price adjustment time
SATURATION_MARKET_PRICE = 97.8  # Saturation market price of electricity

# Assume the Target electric price for demonstration purposes (you can replace it with a dynamic formula)
def target_electric_price(time):
    # For simplicity, let's assume a linear increase in the target electric price
    return 30 + 0.2 * time  # Replace with your actual target price equation

# Function to calculate the indicated electric price
def indicated_electric_price(time):
    target_price = target_electric_price(time)
    return min(SATURATION_MARKET_PRICE, target_price)

# Function to calculate the change in electric price
def change_in_electric_price(busbar_price, time):
    indicated_price = indicated_electric_price(time)
    return (indicated_price - busbar_price) / PRICE_ADJUSTMENT_TIME

# Function to integrate the change in electric price over time to update busbar electricity price
def update_busbar_price(time_steps):
    busbar_price = REFERENCE_BUSBAR_ELEC_PRICE
    busbar_prices = [busbar_price]

    for i, t in enumerate(time_steps[1:], start=1):  # Using enumerate to get index
        change = change_in_electric_price(busbar_price, t)
        # Update busbar price based on the change (we assume a simple Euler method for integration)
        delta_t = t - time_steps[i-1]
        busbar_price += change * delta_t
        busbar_prices.append(busbar_price)

    return np.array(busbar_prices)

# Time steps for simulation (e.g., from 0 to 50 in steps of 0.1)
time_steps = np.arange(0, 50, 0.1)

# Calculate busbar prices over time
busbar_prices = update_busbar_price(time_steps)

# Plot the results
plt.plot(time_steps, busbar_prices, label='Busbar Electricity Price')
plt.xlabel('Time')
plt.ylabel('Busbar Price (in $)')
plt.title('Busbar Electricity Price Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
